```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 1000)
I = np.sin(x) * np.cos(x[:, np.newaxis])
plt.imshow(I)
plt.colorbar();
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
ax=plt.axes(projection='3d')
```

```python
ax=plt.axes(projection='3d')
zline=np.linspace(0,20,100)
xline=np.sin(zline)
yline=np.cos(zline)
ax.plot3D(zline,xline,yline,'g^');
```

```python
ax=plt.axes(projection='3d')
# Data for a three-dimensional line
zline=np.linspace(0,20,100)
xline=np.sin(zline)
yline=np.cos(zline)
ax.plot3D(xline,yline,zline,'r')

# Data for three-dimensional scattered points
zdata=20*np.random.random(100)
xdata=np.sin(zdata)+0.1*np.random.randn(100)
ydata=np.cos(zdata)+0.1*np.random.randn(100)
ax.scatter3D(xdata,ydata,zdata,cmap='inferno');
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
def f(x, y):
    return np.sqrt(16-(x)**2-(y)**2)
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z ,500, cmap='inferno')
ax.contour3D(-X, -Y, -Z ,500, cmap='inferno');
```
