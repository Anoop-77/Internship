```python
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0,10,0.5)
plt.axis([0,20,0,20])

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0,10,0.5)
plt.axis([0,20,0,20])
plt.subplot(1,2,2)
# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
plt.subplot(2,2,1)
# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--')
plt.subplot(2,2,2)
# red dashes, blue squares and green triangles
plt.plot(t, t**3, 'g^')

plt.subplot(2,2,3)
plt.plot(t,t**2,'bs')
plt.show()
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5,
            cmap='viridis')
plt.colorbar();  # show color scale
```

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.randn(1000)
plt.hist(data);
```

```python
plt.hist(data, bins=20, normed=True, alpha=1,   #bins = no. of bars
         histtype='bar', color='red',
         edgecolor='yellow');
```

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(histtype='stepfilled', alpha=0.9, normed=True, bins=40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
```

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T  #inbuilt function ..can be calculated using formula

plt.hist2d(x,y,bins=30,cmap="plasma")
cb=plt.colorbar()
cb.set_label('counts in bin')
```

```python
x=np.linspace(0,10,1000)
plt.plot(x,np.sin(x),'-b',label='Sine')
plt.plot(x,np.cos(x),'r--',label='Cos')
leg=plt.legend()
plt.grid(True)
plt.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1)
plt.show()
```

```python
x=np.linspace(0,10,1000)
y=x+1
z=y+1
plt.plot(x,np.sin(x),'-b',label='Sine')
plt.plot(x,np.sin(y),'r--',label='Cos')
plt.plot(x,np.sin(z),'-y')
leg=plt.legend()
plt.grid(True)
plt.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1)
plt.show()
```
